```python
#importing necessary Libraries
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
import requests
import lxml.html as lh
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import urllib
from selenium.webdriver.chrome.options import Options
import time
```

```python
timeout=30
options = Options()
options.add_argument("window-size=1400,600")
from fake_useragent import UserAgent
ua = UserAgent()
a = ua.random
user_agent = ua.random
options.add_argument(f'user-agent={user_agent}')
driver = webdriver.Firefox()
driver.get('https://www.zomato.com/jaipur/restaurants')
driver.set_page_load_timeout(timeout)
html=driver.page_source
soup=BeautifulSoup(html)
top=soup.find_all(class_="col-s-12")
data=[]
for i in range(0,len(top)):
    link_text=top[i].find(class_='result-title hover_feedback zred bold ln24 fontsize0').text.strip()
    driver.find_element_by_link_text(link_text).click()
    time.sleep(3)
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(3)

    while True:
        try:
   
            loadMoreButton = driver.find_element_by_xpath('//*[@id="reviews-container"]/div[1]/div[3]/div[2]/div')
            loadMoreButton.location_once_scrolled_into_view
            time.sleep(2)
            loadMoreButton.click()
            time.sleep(5)
        except Exception as e:
            break
    

    soup2=BeautifulSoup(driver.page_source)
    text=[]
    text.append(link_text)
    for i in range(0,len(soup2.find_all(class_ ='rev-text mbot0'))):
        text.append(soup2.find_all(class_ ='rev-text mbot0')[i].text)
        text.append('----------------------------------')
    data.append(text)
    driver.get('https://www.zomato.com/jaipur/restaurants')

driver.quit()
```
